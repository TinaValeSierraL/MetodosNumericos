##### Creando nuevos ambientes.

*Es importante crear nuevos ambientes de trabajo, ya que podemos equivocarnos, instalar algo incorrecto o no compatible y asi dejar Base quieto, qué es base? El root de anaconda.*

Que ventajas tenemos de trabajar en ambientes?
   * Activar y desactivarlos
   * Instalar atraves de consola las cosas especificas con su respectiva version.
   

    

#### COMANDOS IMPORTANTES

* conda create -n NombreAmbiente.
* conda activate NombreAmbiente.
* conda deactivate.
* conda env list.
* conda install python=3.5.
* conda instal NombreDelPaquete
